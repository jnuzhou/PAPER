```python
def apply_pca(df, n_components=None, explained_variance_threshold=0.95, plot_variance=True, plot_loadings=True):
    """
    对数据进行主成分分析 (PCA) 降维，并分析每个主成分的特征贡献

    参数：
    - df: DataFrame, 需要进行 PCA 的数据
    - n_components: int, 降维后的特征数（如果 None，则自动选择）
    - explained_variance_threshold: float, 解释方差比例阈值（仅当 n_components=None 时生效）
    - plot_variance: bool, 是否绘制解释方差图
    - plot_loadings: bool, 是否绘制主成分的特征贡献图

    返回：
    - reduced_df: DataFrame, 降维后的数据
    - pca_model: 训练好的 PCA 模型（可用于逆变换或进一步分析）
    - loadings: DataFrame, 载荷矩阵（原始特征对主成分的贡献）
    """
#     # 1. 标准化数据
#     scaler = StandardScaler()
#     df_scaled = scaler.fit_transform(df)

    # 2. 训练 PCA
    df_scaled = df
    pca = PCA(n_components=n_components)
    pca.fit(df_scaled)

    # 3. 计算累积解释方差比例
    cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

    # 4. 如果未指定 n_components，则自动选择解释方差 > threshold 的最小主成分数
    if n_components is None:
        n_components = np.argmax(cumulative_variance >= explained_variance_threshold) + 1
        print(f"自动选择主成分数: {n_components}，累积解释方差: {cumulative_variance[n_components-1]:.4f}")

    # 5. 重新训练 PCA，使用最终的 n
    pca = PCA(n_components=n_components)
    df_pca = pca.fit_transform(df_scaled)

    # 6. 计算载荷矩阵
    loadings = pd.DataFrame(pca.components_.T, 
                            index=df.columns, 
                            columns=[f"PC{i+1}" for i in range(n_components)])

    # 7. 绘制解释方差图
    print(cumulative_variance)
    if plot_variance:
        plt.figure(figsize=(8, 5))
        plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='b')
        plt.axhline(y=explained_variance_threshold, color='r', linestyle='--', label=f'阈值 {explained_variance_threshold}')
        plt.xlabel('主成分数')
        plt.ylabel('累积解释方差')
        plt.title('主成分分析（PCA）解释方差比')
        plt.legend()
        plt.grid(True)
        plt.show()

    # 8. 绘制主成分的特征贡献
    if plot_loadings:
        for i in range(n_components):
            plt.figure(figsize=(10, 5))
            loadings.iloc[:, i].plot(kind="bar", color="steelblue", alpha=0.8)
            plt.axhline(0, color='black', linewidth=1)
            plt.title(f"PC{i+1} 主成分的特征贡献")
            plt.xlabel("特征")
            plt.ylabel("贡献系数")
            plt.show()


    reduced_df = pd.DataFrame(df_pca, columns=[f"PC{i+1}" for i in range(n_components)])

    return reduced_df, pca, loadings

pca_result, pca_model, pca_loadings = apply_pca(X_train, n_components=None, plot_variance=True, plot_loadings=True)
```

# XGB

```python
from optuna.samplers import RandomSampler, CmaEsSampler
def optuna_fit(objective,model_name,sqlite_db,study_name,n_trials=100,direction='maximize'):
#     Optuna 超参数优化

    study = optuna.create_study(
        storage=sqlite_db,
        study_name=study_name,
        sampler=RandomSampler(),
        direction=direction,
        load_if_exists=True
    )

    study.optimize(objective, n_trials)

    # 输出优化结果
    print(f"Best optimized accuracy: {study.best_value:.5f}")
    print(f"Best hyperparameters: {study.best_params}")

    best_params = study.best_params

def fit_model(model_name,best_params):
    # 初始化模型并传入超参数
    model = model_name(**best_params)

    # 训练模型
    model.fit(X_train, y_train)

    # 预测测试集
    y_pred = model.predict(X_test)

    # 输出分类报告和准确率
    print("分类报告：\n", classification_report(y_test, y_pred))
    print("准确率：", accuracy_score(y_test, y_pred))

    # 提取特征重要性
    feature_importances = model.feature_importances_
    features = X_train.columns


    importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': feature_importances
        }).sort_values(by='Importance', ascending=False)


    # 可视化特征重要性
    plt.figure(figsize=(10, 20))
    sns.barplot(
        data=importance_df, 
        x='Importance', 
        y='Feature', 
        hue='Feature',
        palette='viridis',
        legend=False  
    )
    plt.show()
    plt.tight_layout()
```

```python
def objective(trial):
    params = {
        'grow_policy': trial.suggest_categorical('grow_policy', ["depthwise", "lossguide"]),
        'n_estimators': trial.suggest_int('n_estimators', 400, 2000),  # 增加 n_estimators 范围
        'learning_rate': trial.suggest_float('learning_rate', 0.006, 0.6, log=True),  # 降低学习率范围
        'gamma': trial.suggest_float('gamma', 0, 0.5),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.3, 1.0),  # 修改该范围
        'max_depth': trial.suggest_int('max_depth', 3, 15),  # 增加深度范围
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-9, 100.0, log=True),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-9, 100.0, log=True),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1.0, 5.0),  # 处理类别不平衡时有用  
    }

    # 设置 XGBoost 的多分类参数
    params["booster"] = "gbtree"
#     params["objective"] = "multi:softmax"  # 多分类任务
#     params['num_class'] = len(y_train.unique())  # 自动设置类别数量    
    params["objective"] = "binary:logistic"  # 二分类
    params["tree_method"] = "hist"
    params["verbosity"] = 0  

    # 创建交叉验证对象
    cv = StratifiedKFold(n_splits=8, shuffle=True, random_state=42)
#     for train_idx, val_idx in cv.split(X_train, y_train):
#         print(f"Train indices: {len(train_idx)}, Val indices: {len(val_idx)}")

    # 定义评分器
    accuracy_scorer = make_scorer(accuracy_score)

    model = XGBClassifier(**params)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring=accuracy_scorer, n_jobs=-1)

    return scores.mean()

optuna_fit(objective,XGBClassifier,'sqlite:///xgb.db','xgb_class_39_features_to_18_coms',n_trials=1,direction='maximize')
```

# 随机森林

```python
def objective(trial):
    # 定义随机森林的超参数搜索空间
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),  # 决策树数量
        'max_depth': trial.suggest_int('max_depth', 2, 20),  # 树的最大深度
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),  # 最小分割样本数
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),  # 叶节点最少样本数
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),  # 特征选择方式
        'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy']),  # 切分标准
        'max_samples': trial.suggest_float('max_samples', 0.5, 1.0),  # 每棵树的样本数
        'class_weight': trial.suggest_categorical('class_weight', [None, 'balanced']),  # 类别不平衡处理
        'warm_start': trial.suggest_categorical('warm_start', [True, False]),  # 是否重用前次计算的树
    }


    # 创建交叉验证对象
    cv = StratifiedKFold(n_splits=8, shuffle=True, random_state=42)

    # 定义评分器
    accuracy_scorer = make_scorer(accuracy_score)

    # 使用 RandomForestClassifier 进行模型训练
    model = RandomForestClassifier(random_state=42, **params)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring=accuracy_scorer, n_jobs=-1)

    return scores.mean()


optuna_fit(objective,RandomForestClassifier,'sqlite:///rf.db','rf_class_39_features_to_18_coms',n_trials=50,direction='maximize')
```
