```python
df = pd.read_csv('F:\Graduate materials\论文\代码\output_combined.csv')
```

```python
df['商品综合信息'] = df[['商品名称', '商品分类', '商品品牌']].fillna('').astype(str).agg(' '.join, axis=1)
```

```python
empty_or_nan_or_count = (
    df['商品综合信息'].isna() | 
    (df['商品综合信息'] == "") | 
    df['商品综合信息'].apply(lambda x: isinstance(x, (int, float)))
)
print(f"空字符或 NaN 或数字的数量: {empty_or_nan_or_count.sum()}")
```

```python
df = df[~ (df['商品综合信息'].isna() | (df['商品综合信息'] == ""))]
```

```python
fenlei= df['商品综合信息'].apply(lambda x:re.sub('[^\u4E00-\u9FD5]+','',x))
df_fenlei = pd.DataFrame(fenlei)
```

```python
# 加载停用词
with open('stop_words.txt', 'r', encoding='gb18030') as f:
    chinese_stop_words = [line.strip() for line in f.readlines()]
    
def tokenize_with_stopwords(text):
    words = jieba.cut(text)
    return ' '.join(word for word in words if word not in chinese_stop_words)

df_fenlei['tokenize_without_stopwords'] = df_fenlei['商品综合信息'].apply(tokenize_with_stopwords)
df_fenlei
```

```python
texts = df_fenlei['tokenize_without_stopwords'].tolist()

# 实例化Tokenizer，这里的num_words参数可以根据需要设置为词汇表的最大大小
tokenizer = Tokenizer()

# 使用你的文本数据训练tokenizer
tokenizer.fit_on_texts(texts)

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

# 将文本转换为序列
sequences = tokenizer.texts_to_sequences(texts)

# 打印出转换后的序列
print(sequences[0:5])

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
print(len(texts))
print(len(sequences))
tokenizer.index_word.get(10)
word_index = tokenizer.word_index

for word, index in word_index.items():
    print(f"Word: '{word}', Index: {index}")
    if index == 10:  # 假设我们只想看到前10个词
        break
```

```python
# np.mean(df_fenlei['商品综合信息'].apply(lambda x:len([word for word in jieba.cut(x)])))
# df_fenlei['商品综合信息'].apply(lambda x:len([word for word in jieba.cut(x)])).value_counts()
```

```python
max_sequence_length = 8

padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length)

# 加载预训练的词向量
word_vectors = KeyedVectors.load_word2vec_format('F:\\Graduate materials\\Graduate materials\\deep_learning\\deep_learning_doc\\数据集\\chinese_embedding\\chinese_wiki_embeding20000.txt', binary=False)
import numpy as np
embedding_matrix = np.zeros((len(tokenizer.word_index) + 1, word_vectors.vector_size))
for word, i in tokenizer.word_index.items():
    if word in word_vectors.key_to_index:
        embedding_matrix[i] = word_vectors[word]
```

```python
vectorized_sequences = np.zeros((len(padded_sequences), max_sequence_length, word_vectors.vector_size))

for i, sequence in enumerate(padded_sequences):
    for j, index in enumerate(sequence):
        word = tokenizer.index_word.get(index)
        if word in word_vectors.key_to_index:
            vectorized_sequences[i, j, :] = word_vectors[word]
```

# 确定分类数

```python
from sklearn.metrics.pairwise import cosine_similarity
print(vectorized_sequences_load.shape)
sentence_embeddings = np.mean(vectorized_sequences_load, axis=1)  # 结果形状: (24348, 300)

print(sentence_embeddings.shape)
```

```python
from sklearn.cluster import KMeans

SSE = []
K_range = range(3, 12)  # 选择从1到10个聚类
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=8)
    kmeans.fit(sentence_embeddings)
    SSE.append(kmeans.inertia_)  # inertia_ 是 KMeans 的 SSE

# 绘制肘部法则图
plt.figure(figsize=(8, 5))
plt.plot(K_range, SSE, marker='o')
plt.xlabel('聚类数量 (k)')
plt.ylabel('SSE')
plt.title('肘部法则确定聚类数量')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

best_k = 0
best_score = -1

silhouette = []

K_range = range(4, 12)

for k in K_range:  # 尝试从2到9类
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=8)
    kmeans.fit(sentence_embeddings)
    labels = kmeans.labels_
    score = silhouette_score(sentence_embeddings, labels)
    silhouette.append(score)
    
    if score > best_score:
        best_k = k
        best_score = score

print(f"最佳类别数量：{best_k}, 轮廓系数：{best_score:.4f}")

# 绘制肘部法则图
plt.figure(figsize=(8, 5))
plt.plot(K_range, silhouette, marker='o')
plt.xlabel('聚类数量 (k)')
plt.ylabel('轮廓系数')
plt.title('肘部法则确定聚类数量')

save_path = f'..\wiki_embeding_cluster\轮廓系数.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"轮廓系数 图像已保存至 {save_path}")

plt.show()
```

## 类别建模分析

```python
df = pd.read_csv('F:\Graduate materials\论文\代码\output_wiki_vectorized_sequences.csv')
```

```python
df = df.drop(columns = ['商品名称',  '关联店铺', '商品分类', '商品品牌', '浏览量', '视频数','抖音号'])
df['佣金率'] =df['佣金率'].str.replace('%', '').astype(float) / 100
df = df.dropna()
```

```python
df.shape
```

```python
def choose(x):
    if x >= 3:
        return 3 
    else:
        return x
    
df['转化率_1'] = df['转化率_编码'].apply(lambda x: choose(x))
df
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd

# 假设 'cluster_label' 已经在 DataFrame 中
cluster_labels = df['cluster_label'].unique()
metrics_data = []  # 用于存储各类指标
metrics_data1 = []


plt.figure(figsize=(16, 16))
k = 1

# 遍历每个 cluster_label
for cluster in cluster_labels:
    print(f"\n=== 训练 cluster_label = {cluster} 的模型 ===")
    
    # 按 cluster_label 筛选数据
    cluster_data = df[df['cluster_label'] == cluster]
    X = cluster_data.drop(columns=['转化率_编码', '转化率_1', 'cluster_label'])
    y = cluster_data['转化率_1']
    
    # 拆分训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
    
    # 建立随机森林模型
    model = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        bootstrap=True,
        class_weight='balanced',
        max_features='sqrt'
    )
    model.fit(X_train, y_train)
    
    # 预测测试集
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    
    # 获取分类报告
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # 打印完整的分类报告
    print(f"\n分类报告 for cluster_label {cluster}:\n")
    print(classification_report(y_test, y_pred))
    print(f"准确率 for cluster_label {cluster}:", accuracy)

    # 提取特征重要性
    feature_importances = model.feature_importances_
    features = X.columns
    
    importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importances
    })
    
    plt.subplot(2,2,k)
    k += 1
    
    sns.barplot(
    data=importance_df, 
    x='Importance', 
    y='Feature', 
    hue='Feature',
    palette='viridis'
        
    )
    
    plt.xlabel('特征重要性')
    plt.title(f'类别{cluster}_随机森林特征重要性排序')
    
    
    
    
    # 提取 precision, recall, f1-score 数据
    for label in report.keys():
        if label not in ['accuracy','macro avg',"weighted avg"]:
             metrics_data.append({
                "cluster_label": cluster,
                "conversation_level": label,
                "precision": report[label]["precision"],
                "recall": report[label]["recall"],
                "f1-score": report[label]["f1-score"]
            })
    
    
    for label in report.keys():
        if label  in ["weighted avg"]:
             metrics_data1.append({
                "cluster_label": cluster,
                
                "precision": report[label]["precision"],
                "recall": report[label]["recall"],
                "f1-score": report[label]["f1-score"]
            })
    
    

# 将数据转换为 DataFrame
metrics_df = pd.DataFrame(metrics_data)
metrics_df1 = pd.DataFrame(metrics_data1)

# 调整子图布局
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 设置画布大小
plt.figure(figsize=(16, 12))

# 遍历每个指标
for k, metric in enumerate(["precision", "recall", "f1-score"], start=1):
    plt.subplot(2, 2, k)  # 2 行 2 列子图，第 k 个子图
    
    # 使用 Seaborn 绘制分组柱状图
    sns.barplot(
        data=metrics_df,
        x="cluster_label",  # 类别作为 x 轴
        y=metric,  # 指标值作为 y 轴
        hue="conversation_level",  # 分组依据
        palette="viridis",
        alpha=0.7
    )
    
    # 设置子图标题和标签
    plt.title(f"{metric.capitalize()} for Each Class and Cluster", fontsize=14)
    plt.xlabel("cluster_label", fontsize=12)
    plt.ylabel(metric.capitalize(), fontsize=12)
    plt.legend(title='Conversion rate', bbox_to_anchor=(0, 1), loc='upper right')
    plt.xticks(rotation=0)  # 调整 x 轴文字旋转角度以避免重叠
    
# 调整子图布局
plt.tight_layout()

# 显示完整图表
plt.show()
```
